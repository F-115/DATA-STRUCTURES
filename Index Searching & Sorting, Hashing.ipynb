```python
# Key index Searching & Sorting



class KeyIndex:
    def __init__(self, a):

        min = a[0]
        max = a[0]

        original_a = [0]*len(a)

        for i in range(len(a)):
            original_a[i] = a[i]

        for i in a:
            if i > max:
                max = i
            elif i < min:
                min = i

        #print(max, min)
        if min < 0:
            x = min* -1
            max = max + x
            for i in range(len(a)):
                a[i] = (a[i]) + x

        #print(a)

        # two instance variable
        self.k = [0] * (max + 1)  # max = 9
        self.instVar = [a, original_a, min, max]

        for j in range(len(a)):
            self.k[a[j]] = (self.k[a[j]]) + 1

        #print("\nk")
        #print(self.k)

        #print(a, original_a, min)
        #print(self.instVar)


    def searching(self, value):

        minimum = self.instVar[2]
        maximum = self.instVar[3]

        if minimum < 0:
            x = minimum* -1
            value = value + x

        if value > maximum:
            return False

        if self.k[value] != 0:
            return True

        else:
            return False


    def KeyIndexSorting(self):

        a = self.instVar[0]
        original_a = self.instVar[1]
        minimum = self.instVar[2]

        #print(a, original_a, minimum)



        # Sorting ===============
        output = [0]*len(a)

        # output index list
        for i in range(1, len(self.k)):
            self.k[i] = self.k[i] + self.k[i - 1]

        #print("\nUpdated k")
        #print(self.k)

        # output
        for i in range(len(a)-1, -1, -1):
            var = (self.k[a[i]]) = (self.k[a[i]]) - 1

            if minimum < 0:
                output[var] = (a[i]) + minimum
            else:
                output[var] = a[i]

        return output


# Tester
arr = [1,0,6,9,1,0]
#arr = [-1,-2, -3, -5]

test = KeyIndex(arr)
print(test.searching(5))
print(test.KeyIndexSorting())
print()

# outputs
# False
# [0, 0, 1, 1, 6, 9]










### Solution of Task 02 ###

# Hashing




def HashTable(array):

    hashtable = [0]*9

    vowels = ["A", "E", "I", "O", "U"]

    for i in array:
        sum = 0
        consonant = 0

        for j in i:
            if j not in vowels and not (48 <= ord(j) <= 57):
                consonant = consonant + 1

            elif 48 <= ord(j) <= 57 :
                sum = sum + int(j)


        index = ((consonant * 24) + sum) % 9
        #print(i, " ; ", index)

        if hashtable[index] == 0:
            hashtable[index] = i

        else:
            for k in range(index + 1, len(hashtable)):
                if hashtable[k] == 0:
                    hashtable[k] = i
                    break
                index = (index + 1) % len(hashtable)

    return hashtable


# Taster
str_lst = ['ST189B832','ST189B832','TAMIM1165','MARAAM1953','TAJWAR16386']
print(HashTable(str_lst))
```
