```python
  ##################################
  ########## Linear Arrays #########
  ##################################


### Solution of Task 01 ###
### Shift Left k Cells  ###


def leftShift(source, k):

    index = k
    idx = 0

    while index < len(source):

        source[idx] = source[index]

        index = index + 1
        idx = idx + 1

    while idx < len(source):

        source[idx] = 0
        idx = idx + 1

source = [10,20,30,40,50,60]

leftShift(source,3)
print(source)







### Solution of Task 02 ###
### Rotate Left k cells ###


def rotateLeft(source,k):

    index = k
    idx = 0

    first = []

    for i in range(k):

        first.append(source[i])

    while index<len(source):

        source[idx] = source[index]

        index = index + 1
        idx = idx + 1


    j = 0

    while idx <len(source):

        source[idx] = first[j]

        j = j + 1
        idx = idx + 1


source = [10,20,30,40,50,60]

rotateLeft(source,3)

print(source)







### Solution of Task 03 ###
### Shift Right k Cells  ###


source = [10,20,30,40,50,60]

def rightshift(source,k):

  for j in range (0,k):

    for i in range (len(source)-1,0,-1):
      source[i]=source[i-1]

    source[0]=0

  print(source)


rightshift(source,3)







### Solution of Task 04 ###
### Rotate Right k cells ###


source = [10,20,30,40,50,60]

def rightrotate (source,k):
  first = []
  a = len(source) - k

  while a <= len(source)-1 :
    first = first + [source[a]]
    a += 1

  for i in range (len(source)-1,0,-1):
    source[i] = source[i-k]

  j = 0

  while j < len(first):
    source[j] = first[j]
    j += 1

rightrotate(source,3)

print(source)







### Solution of Task 05 ###
### Remove an element from an array ###


def remove(source, size, element):

    for idx in range(element, len(source)):

        if idx != len(source) - 1:
            source[idx] = source[idx + 1]

        elif idx == len(source) - 1:
            source[idx] = 0


source = [10,20,30,40,50,0,0]

remove(source,5,2)

print(source)







### Solution of Task 06 ###
### Remove all occurrences of a particular element from an array ###


def removeAll(source, size, element):

    count = 0

    i = 0
    while i < len(source):
        if source[i] == element:
            j = i
            while j<len(source)-1:
                source[j] = source[j + 1]
                j = j + 1
            i = i - 1
        i = i + 1



source = [10,2,30,2,50,2,2,0,0]

removeAll(source,7,2)

print(source)







### Solution of Task 07 ###
### Splitting an Array ###


def splittingArray(array_A):

    sum1 = 0
    sum2 = 0

    output = False

    for val1 in range(len(array_A)):
        sum1 = sum1 + array_A[val1]

        for val2 in range(val1 + 1, len(array_A)):
            sum2 = sum2 + array_A[val2]

        if sum2 == sum1:
            output = True
            break

        else:

            sum2 = 0

    return output

array_A = [10, 3, 1, 2, 10]

print(splittingArray(array_A))








### Solution of Task 08 ###
### Array series ###


def arraySeries(n):

    array = [0]*n*n

    for group in range(1, n + 1):

        for value in range(1, group + 1):

            index = group*n-value
            array[index] = value

    return array

print(arraySeries(3))







### Solution of Task 09 ###
### Max Bunch Count ###


def maxBunch(array):

    total = 1

    maxBunch = 0

    for idx in range(len(array) - 1):

        if array[idx] == array[idx + 1]:
            total = total + 1

            if total > maxBunch:
                maxBunch = total
        else:
            total = 1

    return maxBunch

print(maxBunch([1,2,2,3,4,4,4]))








### Solution of Task 10 ###
### Repetition ###


def repetition(array):

    count = 1

    repetNum = 0

    output = False

    temp = [0]*len(array)

    countList = []

    for i in range(len(array)):

        if temp[i] == 0:

            for j in range(i+1, len(array)):

                if array[i] == array[j]:

                    temp[j] = 1
                    count = count + 1

            temp[i] = 1
            if count>1:
                countList.append(count)

            count = 1

    for k in range(len(countList)-1):

        for l in range(k+1 ,len(countList)):

            if countList[l] == countList[k]:
                output = True


    return output

print(repetition([4,5,6,6,4,3,6,4]))

print(repetition([3,4,6,3,4,7,4,6,8,6,6]))









```
