```python
# Recursive Selection Sort

def selection_sorting(array, idx ):

    if idx == len(array):
        return

    else:
        min = idx
        j = idx + 1

        while j < len(array):
            if array[min] > array[j]:
                min = j
            j = j + 1


        temp = array[idx]
        array[idx] = array[min]
        array[min] = temp

        sorting(array, idx + 1)

    return array

lst = [10,1,20,3,6,2,5,11,15,2,12,14,17,18,29]

lst = selection_sorting(lst, idx = 0)

print(lst)







### Solution of Task 02 ###


# Recursive Insertion Sort

def insertion_sorting(array, i = 0):

    if i == len(array):
        return

    if i < len(array):

        j = i - 1

        key = array[i]

        while j >= 0 and key < array[j]:
            array[j + 1] = array[j]

            j = j - 1

        array[j + 1] = key

        insertion_sorting(array, i + 1)

    return array

lst = [10,1,20,3,6,2,5,11,15,2,12,14,17,18]

lst = insertion_sorting(lst)

print(lst)







### Solution of Task 03 & 04 ###


# Singly Linked List Bubble & Selection Sorting

class Node:
    def __init__(self, element, n = None):
        self.element = element
        self.next = n

class MyList:

    def __init__(self, array):

        self.head = None

        # ==== Linked list ====
        tail = None
        for data in array:
            itr = Node(data, None)

            if self.head is None:
                self.head = itr
                tail = itr

            else:
                tail.next = itr
                tail = itr

        # ======================

    def showList(self):
        if self.head == None:
            print('Empty List')

        else:
            value = self.head

            while value != None:
                print(value.element, end =' ')
                value = value.next



    # Task 3

    def bubble_sort(self):
        tail = self.head
        while tail != None :
            tail2 = tail.next

            while tail2 != None:
                if tail.element > tail2.element :
                    temp = tail.element
                    tail.element = tail2.element
                    tail2.element = temp

                tail2 = tail2.next
            tail = tail.next



    # Task 4

    def selection_sort(self):
        tail = self.head
        while tail != None:

            min = tail
            var = tail.next

            while var != None:
                if min.element > var.element:
                    min = var
                var = var.next

            temp = tail.element
            tail.element = min.element
            min.element = temp
            tail = tail.next



lst = [10,1,20,3,6,2,5,11,15,2,12,14,17,18]
lst = MyList(lst)
lst.selection_sort()
#lst.bubble_sort()
lst.showList()







### Solution of Task 05 ###


# Doubly List Inseration Sorting

class Node:
    def __init__(self, e, n, p):
        self.element = e
        self.next = n
        self.prev = p


class DoublyList:

    def __init__(self, array):

        self.head = None
        tail = None

        for i in array:
            if self.head == None:
                self.head = Node(i,None, None)
                tail = self.head
            else:
                new = Node(i, None, tail)
                tail.next = new
                tail = tail.next

    def showList(self):
        tail = self.head

        if tail == None:
            print("empty list")

        while tail != None:
            print(tail.element, end=" ")
            tail = tail.next



    # Task 5

    def insertion_sort(self):
        tail = self.head.next
        while tail != None:
            tail2 = tail.prev
            tail3 = tail
            while tail2 != None:
                if tail2.element > tail3.element :
                    temp = tail2.element
                    tail2.element = tail3.element
                    tail3.element = temp
                tail3 = tail2
                tail2 = tail2.prev
            tail = tail.next

lst = [10,1,20,3,0]
lst = DoublyList(lst)
lst.insertion_sort()
lst.showList()






### Solution of Task 06 ###


# Binary Searching

def binary_search(arr, val, R , L = 0):

    if L <= R:

        M = (L + R) // 2

        if val == arr[M]:
            return M

        elif val > arr[M]:
            return binary_search(arr, val , R , M + 1)

        else:
            return binary_search(arr, val, M - 1, L)

    return -1


lst = [2,3,4,5,6]
r = len(lst)-1
print(binary_search(lst, 3, r))








### Solution of Task 07 ###


# Recursive Fibonacci Memorization

import time

A = {}

def F(n):

    if n == 0 or n == 1:
        return 1

    else:
        return F(n - 1) + F(n - 2)

# Febonacci Memorization
def FM(n):

    if n in A:
        return A[n]

    if n == 0 or n == 1:
        return 1

    else:
        value = FM(n - 1) + FM(n - 2)
        A[n] = value
        return value


# -------------Tester----------------
startTime = time.time_ns()
print(FM(45))
duration = time.time_ns()-startTime
print(duration)
```
